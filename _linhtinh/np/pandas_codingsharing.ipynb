```python
# Hàm Cut: Giúp ta phân chia giá trị của một trường liên tục vào những khoảng theo ngưỡng cắt.
# Kết quả trả ra là nhãn của từng khoảng mà chúng ra khai báo
df8 = df.groupby(['Customer ID', 'Segment']).agg({
    'Sales': 'sum'
}).reset_index(drop=False)

#gán bins: high: >1000, normal: 500-100, low: < 500
bins = [-999999, 500, 1000, 999999]
#gán labels
labels = ['low', 'normal', 'high']

df8['Segment_labels'] = pd.cut(df8['Sales'], bins=bins, labels=labels)

df8.head()
```

```python
pd.pivot_table(df8,
               columns=['Segment_labels'],
               index=['Segment'],
               values=['Sales'],
               aggfunc='sum',
               margins=True,
               margins_name='Grand Total')
```

```python
# Hàm qcut() tương tự như cut, tuy nhiên sẽ cắt lớp theo theo % số tương đối
#Note: Có thể không điền q vào, chỉ gán labels cho pandas tự động tính số phần trăm cắt theo tỷ lệ 33%
# q bản chất là số phần trăm
#high: 50%-100
#normal: 25% - 50%
#low: 0-%25%
q = [0, 0.25, 0.5, 1]
labels = ['low', 'normal', 'high']
df8['Segment_labels_qcut'] = pd.qcut(df8['Sales'], q=q, labels=labels)
df8
```

```python
pd.pivot_table(df8,
               columns=['Segment_labels_qcut'],
               index=['Segment'],
               values=['Sales'],
               aggfunc='sum',
               margins=True,
               margins_name='Grand Total')
```

### Merge, Join data

```python
df_sinhvien = pd.DataFrame({
    'ID': ['001', '002', '003', '004', '005'],
    'Name': ['Pham Van Nghia', 'Tong Thuy Linh', 'Le Van Dai', 'Tran Quang Nghia', 'Doan Thu Ha'],
    'Age': [25, 26, 25, 23, 22],
    'Province': ['Nam Dinh', 'Thanh Hoa', 'TP Ho Chi Minh', 'Da Nang', 'Can Tho']
})

df_sinhvien
```

```python
df_score = pd.DataFrame({
    'ID': ['001', '002', '003', '006', '007'],
    'Math': [6.75, 9, 8, 7, 10],
    'Physic': [8, 9, 9, 8.5, 9],
    'Chemistry': [7, 9.5, 7.5, 9, 10],
    'Province': ['Nam Dinh', 'Thanh Hoa', 'TP Ho Chi Minh', 'Quang Nam', 'Nghe An']
})

df_score
```

```python
#merge inner
df_inner = pd.merge(left=df_sinhvien,
                    right=df_score,
                    on=['ID'],
                    how='inner',
                    indicator=True,
                    suffixes=['_sv', '_score'])
df_inner
```

```python
#merge left
df_left = pd.merge(left=df_sinhvien,
                    right=df_score,
                    on=['ID'],
                    how='left',
                    indicator=True,
                    suffixes=['_sv', '_score'])
df_left
```

```python
# Hàm join(): Cẩn thẩn khi dùng join() vì nó chỉ care index của 2 bảng thôi.
df_sinhvien.join(df_score, lsuffix='_sv', rsuffix='_score')
```

```python
# Set lại index cho từng bảng
df_sinhvien.set_index('ID', inplace=True)
df_score.set_index('ID', inplace=True)
#join 2 bảng lại
df_sinhvien.join(df_score, lsuffix='_sv', rsuffix='_score')
```

```python
#Hàm concatenate()
pd.concat([df_sinhvien, df_score],
          axis='rows',
          join='outer',)
```

```python
df_sinhvien.reset_index(inplace=True)
df_score.reset_index(inplace=True)
pd.concat([df_sinhvien.set_index('ID'), df_score.set_index('ID')], axis='columns')
```
