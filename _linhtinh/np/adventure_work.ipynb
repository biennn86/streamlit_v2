#### 1. Import thư viên và dataset

```python
# import thư viện, file

import pandas as pd 
import numpy as np 
import os 
import re 

path = "D:/DATA/P&G/my_project/_linhtinh/np/data_sales/"

lst_file = os.listdir(path=path)
pattern = r'AdventureWorks_Sales'
frames = []
for name_file in lst_file:
    match = re.match(pattern=pattern, string=name_file)
    if match:
        path_file = path + name_file
        df = pd.read_csv(path_file)
        frames.append(df)
df_combine_sales = pd.concat(frames, axis='rows')
df_combine_sales
```

```python
# in tên của file trong folder dataset

for file in lst_file:
    print(path + file)
```

```python
#import các dataset vào dataframe
df_calendar = pd.read_csv('D:/DATA/P&G/my_project/_linhtinh/np/data_sales/AdventureWorks_Calendar.csv')
df_products = pd.read_csv('D:/DATA/P&G/my_project/_linhtinh/np/data_sales/AdventureWorks_Products.csv')
df_pro_cate = pd.read_csv('D:/DATA/P&G/my_project/_linhtinh/np/data_sales/AdventureWorks_Product_Categories.csv')
df_pro_subcate = pd.read_csv('D:/DATA/P&G/my_project/_linhtinh/np/data_sales/AdventureWorks_Product_Subcategories.csv')
df_returns = pd.read_csv('D:/DATA/P&G/my_project/_linhtinh/np/data_sales/AdventureWorks_Returns.csv')
df_rerritories = pd.read_csv('D:/DATA/P&G/my_project/_linhtinh/np/data_sales/AdventureWorks_Territories.csv')
```

#### 2. Exploratory Data

##### Số lượng đơn hàng của từng danh mục sản phẩm

```python
# kiểm tra giá trị của product key ở bảng Sales
df_combine_sales['ProductKey'].unique()
```

```python
#kiểm tra cột product ky trong bảng product
df_products['ProductKey'].unique()
```

```python
#merge của 2 bảng sales và products
df_sale_cate = pd.merge(left=df_combine_sales, right=df_products, on=['ProductKey'], how='left')
```

```python
# kiểm tra giá trị của Product sub-category key
df_sale_cate['ProductSubcategoryKey'].unique()
```

```python
# IN ra 5 dòng đầu tiên của bảng merge ở trên
df_sale_cate.head(5)
```

```python
# merge 2 bảng sale, produc và category
df_sales_sub_cate = pd.merge(left=df_sale_cate,
                             right=df_pro_subcate,
                             left_on=['ProductSubcategoryKey'],
                             right_on=['ProductCategoryKey'],
                             how='left')
df_sales_sub_cate.head()
```

```python
# mege tiếp với bảng territories
df_sales_sub_cate_ter = pd.merge(left=df_sales_sub_cate,
                                 right=df_rerritories,
                                 left_on=['TerritoryKey'],
                                 right_on=['SalesTerritoryKey'],
                                 how='left')
df_sales_sub_cate_ter.head()
```

```python
df_sales_sub_cate_ter_final = pd.merge(left=df_sales_sub_cate_ter,
                                       right=df_pro_cate,
                                       left_on=['ProductCategoryKey'],
                                       right_on=['ProductCategoryKey'],
                                       how='left')
```

```python
# số lượng đơn hàng, số lượng sản phẩm
df_sales_sub_cate_ter_final.groupby(['CategoryName']).agg({
    'OrderQuantity': 'sum',
    'CustomerKey': 'nunique'
}).rename(columns={
    'OrderQuantity': '#Order',
    'CustomerKey': '#Cust'
})
```

```python
# Số lượng đon hàng và số lượng sản phẩm cảu dategory name=Bikes
df_sales_sub_cate_ter_final[df_sales_sub_cate_ter_final['CategoryName'] == 'Bikes'].groupby(['CategoryName']).agg({
    'OrderQuantity': 'sum',
    'CustomerKey': 'nunique'
}).rename(columns={
    'OrderQuantity': '#Order',
    'CustomerKey': '#Cust'
})
```

```python
# Lấy số lượng đơn hàng theo tầng quốc gia
df_sales_sub_cate_ter_final.groupby(['Country']).agg({
    'OrderQuantity': 'sum',
    'CustomerKey': 'nunique'
}).rename(columns={
    'OrderQuantity': '#Order',
    'CustomerKey': '#Cust'
})
```

```python
# Số lượng sản phẩm theo country = France và Germany
df_sales_sub_cate_ter_final[(df_sales_sub_cate_ter_final['Country'] == 'France') | (df_sales_sub_cate_ter_final['Country'] == 'Germany')].groupby(['Country']).agg({
    'OrderQuantity': 'sum',
    'CustomerKey': 'nunique'
}).rename(columns={
    'OrderQuantity': '#Order',
    'CustomerKey': '#Cust'
})
```
